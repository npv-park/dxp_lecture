```python
# 패키지 로딩하기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stat
import seaborn as sns
```

```python
# 엑셀 데이터 불러오기
regData = pd.read_excel("regression.xlsx", sheetname=0)
```

```python
# 데이터의 일부 보기
regData.head(3)
```

```python
# 산점도(Scatter Plot)
plt.scatter(regData["Price_per_week"], regData["Number_of_weekly_riders"])
plt.show()
```

```python
# 산점행렬도(SPM : Scatter Plot Matrix)
sns.pairplot(regData)
plt.show()
```

```python
# 상관계수(Coefficient of Correlation)
regData[["Price_per_week", "Number_of_weekly_riders"]].corr()
```

```python
# 상관계수행렬(Coefficient of Correlation Matrix)
regData.corr()
```

```python
# 상관분석(Correlation Analysis)
stat.pearsonr(regData["Price_per_week"], regData["Number_of_weekly_riders"])
```
